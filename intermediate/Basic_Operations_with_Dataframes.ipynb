```python
#### Basic Operations with Dataframes
```

```python
import pandas as pd

stats = pd.read_csv('/content/sample_data/SampleData.csv')
```

```python
stats.head()
```

```python
stats.columns
```

```python
#stats.describe()
stats.info()
```

```python
# Convert the column from object to int64
stats['Income_USD'] = stats['Income_USD'].astype('int64') # this will give you error if there is ',' comma in column values so use below one.
```

```python
# Remove commas and convert the column to integer
stats['Income_USD'] = stats['Income_USD'].replace({',': ''}, regex=True).astype('int64')
```

```python
stats.columns = ['Age', 'Gender', 'Occupation', 'Income_USD', 'Location', 'MaritalStatus', 'EducationLevel']
```

```python
## Subsetting few columns from above dataframe-
stats[['Age', 'Gender', 'Occupation']] [4:8]
```

```python
## Mathematical Operatoins-
result = stats.Age * stats.Income_USD
result.head()
```

```python
## Add Column into Dataframe-
stats['MyCalc'] = stats.Age * stats.Income_USD
stats
```

```python
## Comparison to R
stats['PQR'] = [1, 2, 3, 4, 5] # Error as Python has no recycling option.
```

```python
## Removing a column
```

```python
stats.head()
```

**Explanation:**
1. 'MyCalc': This is the column you want to drop.
2. axis=1: This indicates that you want to drop a column. (If you want to drop rows, you'd use axis=0).
3. inplace=True: This will modify the DataFrame in place, meaning it will directly update stats. If you want to return a new DataFrame without modifying the original, you can remove inplace=True.

```python
#stats.drop('MyCalc', axis=1, inplace=True)
stats.drop('MyCalc', axis=1)
```

```python
stats.head()
```

```python
stats.drop('MyCalc', axis=1, inplace=True)
```
