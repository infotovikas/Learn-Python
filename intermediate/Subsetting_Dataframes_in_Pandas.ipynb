```python
#### Subsetting Dataframes in Pandas
```

```python
import pandas as pd

stats = pd.read_csv('/content/sample_data/SampleData.csv')
```

```python
stats.head()
```

```python
#Three parts will cover-
# -Rows
# -Columns
#-Combine the above two
```

```python
#Part1. -Rows
stats[2:4]   # slicing the dataframe and getting row from 2 to 4 as its indexing start from 0 so 4th row will be exluded
```

```python
stats[:] #get everything in stats - all the rows
```

```python
stats[3:] # start from 3rd to end
```

```python
stats[:7]  # starts from begining till 7th like top 7 (stats.head(7))
```

```python
## Reverse the dataframe
stats[ : : -1] # (start from begining, till end, step out -1)
```

```python
## Gen only every 3rd rows
stats[::-3]
```
